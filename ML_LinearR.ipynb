```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
df
```

```python
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df.year,df.income, color='green', marker='+')
plt.show()
```

```python
year_df = df[["year"]]
income = df.income
```

#Creating linear regression object


```python
reg = linear_model.LinearRegression()
reg.fit(year_df,income)
```

# y = mx + b
# m = reg.coef_
# b = reg.intercept

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict(pd.DataFrame([[2020]], columns=['year']))
```
